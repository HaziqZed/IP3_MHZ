```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

```python
def to_dataframe(Filename):
    df = pd.read_csv(Filename)
    df.Date = pd.to_datetime(df.Date)
    name = Filename.split('.')
    df.insert(0,"Company",name[0])
    return df
```

```python
df1 = to_dataframe('50Hertz.csv')
df2 = to_dataframe('Amprion.csv')
df3 = to_dataframe('TenneTTSO.csv')
df4 = to_dataframe('TransnetBW.csv')
```

```python
df4
```

```python
df = pd.concat([df1,df2,df3,df4])
```

```python
df.sample(10)
```

```python
ndf = df.groupby([df.Company, df.Date.dt.year]).mean()
```

```python
ndf
```

```python
midnight = ndf['00:00:00']
midnight
```

```python
noon = ndf['12:00:00']
noon
```

```python
ax1 = midnight.sort_values().plot(kind='barh', title='TW/hours by Company/Year @ Midnight')
```

```python
ax2 = noon.sort_values().plot(kind='barh', color='red', title='TW/hours by Company/Year @ Noon')
```

```python
fig, axes = plt.subplots(nrows=2, sharex=True, figsize=(12,6))
axes[0] = midnight.sort_values().plot(kind='barh', title='TW/hours by Company/Year @ Midnight')
axes[1] = noon.sort_values().plot(kind='barh', color='red', title='TW/hours by Company/Year @ Noon')
```
