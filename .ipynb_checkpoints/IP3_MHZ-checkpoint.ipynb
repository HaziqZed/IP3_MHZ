```python
import pandas as pd

import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
%matplotlib inline
```

**1. The function written to read in the files into dataframes.**

```python
def to_dataframe(Filename):
    '''This function reads the filename and puts it into a DataFrame, and then converts the Date column
    to the datetime format. Then it adds a column of the company name derived from the filename'''
    df = pd.read_csv(Filename)
    df.Date = pd.to_datetime(df.Date)
    name = Filename.split('.')
    df.insert(0,"Company",name[0])
    return df
```

**2. Executing the to_dataframe function four times for four different csv files**

```python
df1 = to_dataframe('50Hertz.csv')
df2 = to_dataframe('Amprion.csv')
df3 = to_dataframe('TenneTTSO.csv')
df4 = to_dataframe('TransnetBW.csv')
```

**3. Stacking or Concatinating the four dataframes into one big dataframe**

```python
df = pd.concat([df1,df2,df3,df4])
```

```python
df.sample(10)
```

**4. Using the grouby function to visualize the aggregate data, or the mean for each company at the different times for both 2019 and 2020.**

```python
ndf = df.groupby([df.Company, df.Date.dt.year]).mean()
```

```python
ndf
```

**5. Creating the series for midnight and noon, and then using the plot function to create a bar chart with the values that are sorted.**

```python
midnight = ndf['00:00:00']
midnight
```

```python
noon = ndf['12:00:00']
noon
```

```python
figm = plt.figure(figsize=(12,6))
axm = figm.add_subplot(221)
axm = midnight.sort_values().plot(kind='barh', title='TW/hours by Company/Year @ Midnight')
figm.savefig('midnight.png')
```

```python
fign = plt.figure(figsize=(12,6))
axn = fign.add_subplot(221)
axn = noon.sort_values().plot(kind='barh', color='red', title='TW/hours by Company/Year @ Noon')
fign.savefig('noon.png')
```

**6. Combining the two graphs that were previously created into a single figure where the graphs share the same x-axis and then changing the colors of the bars respectively.**

```python
fig = plt.figure(figsize=(12,6))
ax1 = fig.add_subplot(2,1,1)
ax1 = midnight.sort_values().plot(kind='barh', title='TW/hours by Company/Year @ Midnight',color='silver')
ax2 = fig.add_subplot(2,1,2, sharex=ax1)
ax2 = noon.sort_values().plot(kind='barh', color='grey', title='TW/hours by Company/Year @ Noon')
fig.savefig('combined_bar_chart.png')
```

**7. Creating the function that accepts an ungrouped dataframe, the company name, date, and a plotting color and would return an area graph for that particular company on that specific date.**

```python
def daily_view(dataframe, company, date, color):
    '''This function accepts the dataframe, company, date and color. It creates a figure to plot the graph.
    It takes the company name and date, and creates a new dataframe consiting of the specific data for the 
    given company and date. Then, the function converts the dataframe into a series to be able to plot. Then,
    it plots the graph with the particular color and axis labels, and title.'''
    fig1 = plt.figure(figsize=(12,6))
    new = df[df['Company']==company]
    new1 = new[new['Date']==date]
    new1 = new1.drop(columns=['Company', 'Date'])
    new1 = new1.T.squeeze()
    ax1 = new1.plot(kind='area',color = color, title=f'Power Production: {company}/{date}')
    ax1.set_xlabel('Time of Day')
    ax1.set_ylabel('TW/h Generated')
    return 
```

```python
daily_view(df,'50Hertz','2019-08-24','purple')
```
